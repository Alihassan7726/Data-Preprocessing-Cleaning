```python
df['make'].replace(to_replace='?', value=np.nan , inplace=True)
df['price'].replace(to_replace='?', value=np.nan , inplace=True)
df2 = df[(df['price'].isna())]
df2
```

```python
pd.isna(df['price'])
```

```python
df['price'] = pd.to_numeric(df['price'])
df['price'].describe()
```

```python
data_to_imputeprice = df.groupby(['make','body-style'])['price'].mean()
data_to_imputeprice
```

```python
# Grouping price according to make and body-style
data_to_imputeprice22 = df.groupby(['make','body-style'])['price']
data_to_imputeprice22
```

```python
# Now imputing with mean of prices that were grouped accordingly and also they will impute accordingly 
df['price'] = data_to_imputeprice22.transform(lambda x : x.fillna(x.mean()) )
df['price']
```
